```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
url = "https://raw.githubusercontent.com/dilp90/InteligenciaArtificial_y_RedesNeuronales_UANL_FIME/main/MachineLearning/Datasets/robot_sensors.csv "
df = pd.read_csv(url)

try:
    df = pd.read_csv(url)
    print(df.head())
except pd.errors.ParserError as e:
    print(f"Error reading CSV: {e}")
    # Add any desired error handling logic here
```

```python
df = df.dropna(subset=["Distancia (cm)","Velocidad (cm/s)","Temperatura (°C)"])
print(df)
```

```python
df_filtrado = df[(df["Distancia (cm)"] >0) | (df["Velocidad (cm/s)"] >0) | (df["Temperatura (°C)"])>0]
print(df_filtrado)
```

```python
from sklearn.preprocessing import MinMaxScaler , StandardScaler
df = pd.DataFrame(df)
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
print(df_scaled)

scaler_zscore = StandardScaler()
df_zscore = pd.DataFrame(scaler_zscore.fit_transform(df), columns=df.columns)
print(df_zscore)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(df)
plt.title("Datos originales")
plt.legend(df.columns)

plt.subplot(1,2,2)
plt.plot(df_scaled )
plt.title("Datos Normalizados (Min_Max)")
plt.legend(df.columns)

plt.tight_layout()
plt.show()
```
